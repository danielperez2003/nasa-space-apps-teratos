```python
# Add data objects to weaviate
import json
# Get data from NTRS
data_file = open('../data.json')
data = json.load(data_file)
# Closing file
data_file.close()

# Add data to Vector DB
for document in data['documents']:
    client.batch.add_data_object(document, 'Document')

client.batch.flush()

# Where an example of a data object to be pushed into the database could look like:
"""
{
            "docId": "19700031801", 
            "stiTypeDetails": "Contractor Report (CR)", 
            "title": "Geological evaluation of Nimbus vidicon photography, Chesapeake Bay-Blue Ridge", 
            "abstract": "Geological evaluation of Nimbus vidicon photography of Chesapeake Bay to Blue Ridge area", 
            "keywords": ["topographic", "bedrock", "water bundaries", "triassic"], 
            "submittedDate": "2013-08-05T17:47:00.0000000+00:00", 
            "subjectCategories": ["GEOPHYSICS"], 
            "author": "Davies, W. E."
      
}
"""
```

```python
# Querying data coming from the frontend
import weaviate

client = weaviate.client.Client('http://localhost:8080')
# Sample Inputs
# In this example, the user may have entered a natural language description for the documents
# he/she would like to get. The NLP techniques would have processed the input extracting (in this example)
# some keywords and what could be a category from recognising the intention of the input.
kw = ['map', 'topography']
categories = ['geography']
content = {
    'concepts': kw,
    'certainty': 0.79, # or 'distance' instead
    'moveTo': {
        'concepts': categories,
        'force': 0.85
    }
}
# Here the database will look for the KNN based on cosine similarity of the <content> definition
response = client.query.get('Document', ['author', 'title'])\
    .with_near_text(content)\
    .with_additional(['certainty'])\
    .do()

print(response)

# This would be an example output for this final code
"""
{'data': {
  'Get': {
    'Document': [{'_additional': {'certainty': 0.8362697958946228}, 'author': 'Cooper, J. R.', 'title': 'Geologic evaluation - Radar imagery of Twin Buttes area, Arizona test site 15'}, 
    {'_additional': {'certainty': 0.8052285015583038}, 'author': 'Davies, W. E.', 'title': 'Geological evaluation of Nimbus vidicon photography, Chesapeake Bay-Blue Ridge'}]}}}
"""
```
